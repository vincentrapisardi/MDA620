The dataset contains the following fields:

    key - a unique identifier for each trip

    passenger_count - the number of passengers in the vehicle (driver entered value)
    pickup_longitude - the longitude where the meter was engaged
    pickup_latitude - the latitude where the meter was engaged
    dropoff_longitude - the longitude where the meter was disengaged
    dropoff_latitude - the latitude where the meter was disengaged

#### Converting Data Structure

```python
#pd.to_datetime(uber['pickup_datetime'], format='%Y-%m-%d hh:mm:ss')

uber['pick_date'] = pd.to_datetime(uber['pickup_datetime']).dt.date
```

```python
uber['pick_month']= pd.DatetimeIndex(uber['pick_date']).month
uber['pick_year']= pd.DatetimeIndex(uber['pick_date']).year

uber['pick_month_abb']= pd.DatetimeIndex(uber['pick_date']).month_name()


uber.head(2)
```

```python
uber.groupby('pick_month_abb',sort=True)['fare_amount'].sum()
```

```python
uber.groupby(['pick_month_abb', 'pick_year'])['fare_amount'].sum()
```
